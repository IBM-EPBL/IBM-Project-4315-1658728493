# Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the Dataset

```python
df = pd.read_csv('autos.csv', encoding = 'latin1')
```

```python
df
```

## Cleaning the Dataset

```python
df.drop(columns =['dateCrawled', 'nrOfPictures'], inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df['vehicleType'].fillna(df['vehicleType'].mode()[0], inplace = True)
df['gearbox'].fillna(df['gearbox'].mode()[0], inplace = True)
df['model'].fillna(df['model'].mode()[0], inplace = True)
df['fuelType'].fillna(df['fuelType'].mode()[0], inplace = True)
df['notRepairedDamage'].fillna(df['notRepairedDamage'].mode()[0], inplace = True)
```

```python
df.isna().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['name'] = le.fit_transform(df['name'])
df['seller'] = le.fit_transform(df['seller'])
df['offerType'] = le.fit_transform(df['offerType'])
df['abtest'] = le.fit_transform(df['abtest'])
df['vehicleType'] = le.fit_transform(df['vehicleType'])
df['gearbox'] = le.fit_transform(df['gearbox'])
df['model'] = le.fit_transform(df['model'])
df['fuelType'] = le.fit_transform(df['fuelType'])
df['brand'] = le.fit_transform(df['brand'])
df['notRepairedDamage'] = le.fit_transform(df['notRepairedDamage'])
df['dateCreated'] = le.fit_transform(df['dateCreated'])
df['lastSeen'] = le.fit_transform(df['lastSeen'])
```
